# Python List `clear()`

The **`clear()`** method removes all items from the list.

**Syntax**:

```python
list.clear()
```

## `pop()` Parameters

The **`clear()`** method doesn't take any parameters.

## Return Value from `clear()`

The **`clear()`** method only empties the given **[list](http://localhost:8888/notebooks/01_Learn_Python4Data/02_Python_Datatypes/004_Python_List.ipynb)**. It doesn't return any value.

````python
# Example 1: Working of clear() method

# Defining a list
list = [{1, 2}, ('a'), ['1.1', '2.2']]

# clearing the list
list.clear()

print('List:', list)
````

>**Note:** If you are using Python 2 or Python 3.2 and below, you cannot use the **`clear()`** method. You can use the del operator instead.

````python
# Example 2: Emptying the List Using del

# Defining a list
list = [{1, 2}, ('a'), ['1.1', '2.2']]

# clearing the list
del list[:]

print('List:', list)
````
